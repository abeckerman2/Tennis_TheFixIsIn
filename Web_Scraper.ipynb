```python
import requests
from bs4 import BeautifulSoup

URL = "https://www.oddsshark.com/tennis/atp/odds"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
#name_results = soup.find_all("div", {"class": "participant-name"})
odds_results = soup.find_all("div", {"class": "odds-moneyline moneyline-only"})
#odds_results = soup.find_all("div", {"class": "data-odds-moneyline"})
name_results = soup.find_all(lambda tag: tag.name == 'div' and 
                                   tag.get('class') == ['participant-name'])

#odds_results = soup.find_all(lambda tag: tag.name == 'div' and 
                                   #tag.get('class') == ['odds-moneyline moneyline-only'])
#odds_results = soup.find_all(lambda tag: tag.name == 'div' and 
#                                   tag.get('class') == ['odds--group__events-container'])
```

```python
list_names = []
for i in name_results:
    #print(i, end="\n"*2)
    print(i.text)
    list_names.append(i.text)
```

```python
import pandas as pd
df = pd.DataFrame()
```

```python
player_one_list = []
player_two_list = []
for index, row in enumerate(list_names):
    if index % 2 == 0:
        player_one_list.append(row)
    if index % 2 != 0:
        player_two_list.append(row)
print(player_one_list)
print(player_two_list)
```

```python
df['player_one'] = player_one_list
df['player_two'] = player_two_list

df
```

```python
#odds_results
list_odds = []
for i in odds_results:
    print(i.text, end="\n"*2)
    list_odds.append(i.text)
```

```python
len(list_odds)
```

```python
current_odd_list = [v for i, v in enumerate(list_odds) if i % 4 < 2]

player_one_list = []
player_two_list = []
for index, row in enumerate(current_odd_list):
    if index % 2 == 0:
        player_one_list.append(row)
    if index % 2 != 0:
        player_two_list.append(row)
print(player_one_list)
print(player_two_list)


df['player_one_odds'] = player_one_list
df['player_two_odds'] = player_two_list

df
```
